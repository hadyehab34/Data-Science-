# Cleaning

### 1) Quality

#### Starting the clean dataset with no rows with more than 10 missing values

```python
movies_clean = movies.copy()
movies_clean = movies_clean[movies_clean.isnull().sum(axis=1) < 10]
```

```python
movies_clean
```

#### Filling some columns null values with every its average value

```python
avg_1 = movies_clean.num_critic_for_reviews.sum()/len(movies.columns)
avg_2 = movies_clean.num_user_for_reviews.sum()/len(movies.columns)
avg_3 = movies_clean.aspect_ratio.sum()/len(movies.columns)
```

```python
movies_clean['num_critic_for_reviews'] = movies_clean.num_critic_for_reviews.fillna(avg_1)
movies_clean['num_user_for_reviews'] = movies_clean.num_user_for_reviews.fillna(avg_2)
movies_clean['aspect_ratio'] = movies_clean.aspect_ratio.fillna(avg_3)
```

```python
movies_clean.isnull().sum()
```

#### Removing records that have null values in certain coulmns (for classification purposes)

```python
movies_clean = movies_clean[~movies_clean.duration.isnull()]
movies_clean = movies_clean[~movies_clean.gross.isnull()]
movies_clean = movies_clean[~movies_clean.budget.isnull()]
movies_clean = movies_clean[~movies_clean.director_facebook_likes.isnull()]
movies_clean = movies_clean[~movies_clean.actor_3_facebook_likes.isnull()]
movies_clean = movies_clean[~movies_clean.actor_2_facebook_likes.isnull()]
movies_clean = movies_clean[~movies_clean.actor_1_facebook_likes.isnull()]
movies_clean = movies_clean[~movies_clean.cast_total_facebook_likes.isnull()]
movies_clean = movies_clean[~movies_clean.movie_facebook_likes.isnull()]
movies_clean = movies_clean[~movies_clean.facenumber_in_poster.isnull()]
```

### 2) Tidiness

#### Splitting multi-valued columns into 2 columns

```python
movies_clean['plot_keyword_1'] = movies_clean['plot_keywords'].str.split('|').str.get(0)
movies_clean['plot_keyword_2'] = movies_clean['plot_keywords'].str.split('|').str.get(1)
```

```python
movies_clean = movies_clean.drop(['plot_keywords'], axis=1)
```

```python
movies_clean['genre_1'] = movies_clean['genres'].str.split('|').str.get(0)
movies_clean['genre_2'] = movies_clean['genres'].str.split('|').str.get(1)
movies_clean['genre_2'] = movies_clean.genre_2.fillna(movies_clean['genre_1'])
```

# Classification

```python
import numpy as np
conditions = [
    (movies_clean['imdb_score'] >= 7.8),
    ((movies_clean['imdb_score'] < 7.8) & (movies_clean['imdb_score'] >= 5.5)),
    (movies_clean['imdb_score'] < 5.5)]

# create a list of the values we want to assign for each condition
values = ['Class_1', 'Class_2','Class_3']

# create a new column and use np.select to assign values to it using our lists as arguments
movies_clean['IMDB_Category'] = np.select(conditions, values)

# display updated DataFrame
movies_clean.head(10)
```

```python
movies_clean[movies_clean.IMDB_Category == "Class_2"]
```

```python
movies_clean[movies_clean.IMDB_Category == "Class_1"].imdb_score.describe()
```

```python
movies_clean[movies_clean.IMDB_Category == "Class_2"].imdb_score.describe()
```

```python
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier # Import Decision Tree Classifier
from sklearn.model_selection import train_test_split # Import train_test_split function
from sklearn import metrics #Import scikit-learn metrics module for accuracy calculation

X = movies_clean[['gross','budget','title_year','director_facebook_likes','actor_1_facebook_likes','actor_2_facebook_likes','actor_3_facebook_likes','cast_total_facebook_likes','duration','num_critic_for_reviews','num_voted_users']]
Y = movies_clean.IMDB_Category

X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.3, random_state=1) # 70% training and 30% test
```

```python
# Create Decision Tree classifer object
clf = DecisionTreeClassifier()

# Train Decision Tree Classifer
clf = clf.fit(X_train,y_train)

#Predict the response for test dataset
y_pred = clf.predict(X_test)
```

```python
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```
